# Q.2: What proportion of DAU is contributed by new vs existing users?

```python
import math
```

```python
list2=[] #all unique user ids
prop=[]  #final result
for key, value in df.iteritems():
    new_cnt=0
    old_cnt=0
    for keys, values in value.iteritems():
        if math.isnan(values)!=True:
            if values not in list2:
                list2.append(values)
                new_cnt+=1
            else:
                old_cnt+=1
        else: pass
    prop.append([key,old_cnt,new_cnt])

print(prop)



```

```python

#New and existing count for each day
dfx = pd.DataFrame(prop, columns = ['Dates','Existing_count','New_count'])
dfx.head()
```

```python
Total_logins = dfx['Existing_count'].sum()+dfx['New_count'].sum()
print("Total number of user logins is",Total_logins)
print("Total number of Existing user logins is",dfx['Existing_count'].sum())
print("Total number of new user logins is",dfx['New_count'].sum())
```

```python
# A bar plot to show proportion of new vs existing users for each day
dfx.plot.bar(width= .8,figsize=(20,10))
plt.title('Daily Existing v/s New users',size= 20)
plt.xlabel('Days',size=20)
plt.ylabel('Number of users(Existing & New)',size =20)
plt.show()
```

```python
# A pie plot to show proportion of new vs old users in entire dataframe
y = np.array([dfx['New_count'].sum(),dfx['Existing_count'].sum()])
mylabels = ["New_Users","Existing_Users"]
plt.pie(y, labels = mylabels)
plt.show()

New_proportion = (dfx['New_count'].sum()/Total_logins)*100
print("New users proportion:",New_proportion,"percent")
Existing_proportion = (dfx['Existing_count'].sum()/Total_logins)*100
print("Existing Users Proportion",Existing_proportion,"percent")
```

# Q.4: Create a day-on-day retention analysis for every user and derive insights around metrics mentioned above or any others you think are necessary.

```python
import seaborn as sns
```

```python
#creating a group of Signup_day and membership period
group = df6.groupby(['Signup_Day', 'Membership_Period_Days'])

cohort_data = group['ID'].size()

cohort_data = cohort_data.reset_index() 

cohort_data
```

```python
cohort_data['Membership_Period_Days'] = cohort_data['Membership_Period_Days'].dt.days
```

```python
cohort_counts = cohort_data.pivot(index='Signup_Day', columns='Membership_Period_Days', values='ID')

# we just take the first column of cohort_counts as base for retention rate calculation
base = cohort_counts[0]

# And now we apply the divide method to our cohort_counts, dividing it by the base
retention = cohort_counts.divide(base, axis=0).round(3)
cohort_counts
```

```python
retention.fillna(0)
```

# Daily User Retention

```python
plt.figure(figsize=(60,40))
plt.title('Users Active')
ax = sns.heatmap(data=cohort_counts, annot=True, vmin=0.0,cmap='Reds')
ax.set_yticklabels(cohort_counts.index)
fig=ax.get_figure()
fig.savefig("Retention Counts.png")
```

```python
plt.figure(figsize=(60,40))
plt.title('Retention Table')
ax = sns.heatmap(data=retention, annot=True, fmt='.0%', vmin=0.0,cmap='Reds')
ax.set_yticklabels(retention.index)
fig=ax.get_figure()
fig.savefig("Retention_Rate.png")
```

```python
df12.head()
```

```python
df12 = df12.rename(columns={0: "ID", 1: "Cluster"})
```

```python
y = df12.pop('Cluster')
X = df12
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
```

# KNN using elbow method

```python
error_rate = []
for i in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
    
```

```python
Knn_model = KNeighborsClassifier(n_neighbors=5)
Knn_model.fit(X_train, y_train)

pred_1 = Knn_model.predict(X_test)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50), error_rate, color='green', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=5)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

# Metrics for KNN

```python
print("Accuracy:\n", metrics.accuracy_score(y_test, pred_1))
print("Confusion Matrix:\n", metrics.confusion_matrix(y_test, pred_1))
print("Classification Report:\n", metrics.classification_report(y_test, pred_1))
```
